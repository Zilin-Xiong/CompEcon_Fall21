```python
import pandas as pd
import numpy as np
import scipy.optimize as opt
import statsmodels.api as sm
```

```python
fulldata = pd.read_stata('C:/Users/PC/Downloads/PS4_data.dta')
fulldata.head(n=10)
```

```python
fulldata.describe()
```

```python
#data cleaning, to find male heads between 25 and 60, wages>7$/hour
fulldata['wph'] = fulldata['hlabinc']/fulldata['hannhrs']
data1 = fulldata[fulldata['wph']>7]
data1 = fulldata[(fulldata['hannhrs']>0)&(fulldata['hsex']==1)&(fulldata['age']>25)&(fulldata['age']<60)]
data1.head(n=10)
```

```python
#creating dummies variables for races, because we only have white, black and native.
data2 = pd.get_dummies(data1, columns = ['hrace'])
data2.rename(columns={'hrace_1.0':'white','hrace_2.0':'black','hrace_3.0':'otherraces' }, inplace=True)
data2.head()
```

```python
#transformation of wage per hour to log
data2['logwph']=np.log10(data2['wph'])
#create Age^2 variable
data2['AgeSq']=data2['age']*data2['age']
data2.describe()
```

```python
#Select all the variables into a new data frame
data3 = data2[['year','hyrsed', 'age','AgeSq','black','otherraces','logwph']]
data3 = data3.dropna()
data3.describe()
```

```python
#run an OLS to set up a baseline
x = data3[['year','hyrsed', 'age','AgeSq','black','otherraces']]
y = data3['logwph']
ols_model1 = sm.OLS(y,x,missing = 'drop')
results_1 = ols_model1.fit()
results_1.summary()

# Set bounds on parameters
bnds = ((0, 1000), (0, 1000), (0, 1000),
        (None, 1000), (None, 1000), (0, 1000))
```

```python
##write the objective function
data_MLE = {}
dv = {}
min = {}

for n in [1971,1980,1990,2000]:
    data_MLE[n] = data3[data3.year == n]
    dv[n] = data_MLE[n]['logwph']
    def myfunction(parameters):
        '''
        Calculate the log-likelihood function, then we find that max LL = min SSE
        Parameters: A length 6 tuple, model parameters as (b, beta1, beta2, beta3, beta4, beta5)
        Return: the sum value of the error terms, that we need to minimize
        '''
        b = parameters[0]
        beta1 = parameters[1]
        beta2 = parameters[2]
        beta3 = parameters[3]
        beta4 = parameters[4]
        beta5 = parameters[5]
        #predict the value of the wage per hour
        yhat = b + beta1*data_MLE[n]['hyrsed']+beta2*data_MLE[n]['age']+beta3*data_MLE[n]['AgeSq']+beta4*data_MLE[n]['black']
        +beta5*data_MLE[n]['otherraces']
        #Calculate the SSE
        SSE = np.array((dv[n] - yhat)**2).sum()
        return SSE
    #Call the minimizer
    '''
    Here we predict the value of beta based on real-world meanings, for example, one more year of education may lead to 5% increase
        in the wage, so we predict the value of beta as 0.05
    For minority races, we assume they will receive higher wage increase as 0.03.
    '''
    min[n] = opt.minimize(myfunction,np.array([0,0.05,0.05,0,0.03,0]),method = 'L-BFGS-B',bounds = bnds)
    print('Beta1 is',min[n].x[1],'for',n)
    print('Beta2 is',min[n].x[2],'for',n)
    print('Beta3 is',min[n].x[3],'for',n)
    print('Beta4 is',min[n].x[4],'for',n)
    print('Beta5 is',min[n].x[5],'for',n)
```
